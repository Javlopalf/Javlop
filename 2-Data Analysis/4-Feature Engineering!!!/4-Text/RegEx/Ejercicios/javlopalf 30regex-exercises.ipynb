1) Write a Python program to check that a string contains only a certain set of characters (in this case a-z, A-Z and 0-9).

```python
def check(string):
    palabra = re.compile(r'[\w]')
    string = palabra.search(string)
    return  bool(string)

print(check("Hola"))
print(check("%%%%&&"))



   
```

```python
def check(string):
    
    string = re.fullmatch(r'[a-zA-z0-9]+', string)
    return  bool(string)

print(check("Hola"))
print(check("%%%%&&"))
print(check("holaa?lo"))
```

2) Write a Python program that matches a string that has an a followed by zero or more b's

```python
def check_texto(text):
    patterns = 'ab*'
    if re.search(patterns, text):
        return ("Match")
    else:
        return ("Fallaste wey")


print(check_texto("11"))
print(check_texto("abbbc"))
print(check_texto("ddda"))             
        
```

7) Write a Python program to find sequences of lowercase letters joined with a underscore.

```python
def underscore(textaa):
    paternslow = '[^A-Z_.]'
    if re.search(paternslow, textaa):
        return ("match")
    else:
        return("Fallaste wey")

print(underscore("A_a"))
print(underscore("a_a"))
```

8) Write a Python program to find the sequences of one upper case letter followed by lower case letters.

```python
def mix(texocon):
    paternslow = '[A-Z][a-z]+$'
    if re.search(paternslow, texocon):
        return ("match")
    else:
        return("Fallaste wey")

print(mix("Abaaab"))
print(mix("bbb"))
```

9) Write a Python program that matches a string that has an 'a' followed by anything, ending in 'b'.

```python
def withb(textacoo):
        patterns44 = 'a.*b$'
        if re.search(patterns44,  textacoo):
                return (" match")
        else:
                return("fallaste wey!")

print(withb("aa"))
print(withb("aaaabab"))
print(withb("pcoo"))
```

10) Write a Python program that matches a word at the beginning of a string.

```python
def ones(textacoo):
        patterns44 = '[a-zA-z]'
        if re.match(patterns44,  textacoo):
                return (" match")
        else:
                return("fallaste wey")

print(ones("aa"))
print(ones("aaaabab"))

print(ones("111"))
```

11) Write a Python program that matches a word at the end of string, with optional punctuation.

```python
def punt(textacoo):
        patterns44 = '^\w+'
        if re.match(patterns44,  textacoo):
                return (" match")
        else:
                return("fallaste wey")

print(punt("aaa."))
print(punt("aaaabab"))


```
